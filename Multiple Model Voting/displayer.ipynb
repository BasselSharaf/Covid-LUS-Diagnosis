```python
    import seaborn as sn
    import pandas as pd
    import matplotlib.pyplot as plt
sn.set(font_scale=1.4)
Labels=['Covid','Pneumonia','Regular']
array = [[ 2 , 1 , 4],
 [ 1 , 4  ,2],
 [ 0 , 0 ,11]]
df_cm = pd.DataFrame(array, index = [i for i in Labels],
                  columns = [i for i in Labels])
plt.figure(figsize = (6,6))
sn.heatmap(df_cm, annot=True,cmap=plt.cm.gray)
```

```python
print('SVM without balancing')
Labels=['Covid','Pneumonia','Healthy']
array = [[28 , 2 , 0],
 [ 0,33 , 2],
 [ 0, 5, 50]]
df_cm = pd.DataFrame(array, index = [i for i in Labels],
                  columns = [i for i in Labels])
plt.figure(figsize = (6,6))
sn.heatmap(df_cm, annot=True,cmap="YlGnBu")
```

```python
print('SVM with balancing')


array = [[22 , 1 , 5],
 [ 4 ,32 , 5],
 [ 4 , 3 ,46]]
df_cm = pd.DataFrame(array, index = [i for i in Labels],
                  columns = [i for i in Labels])
ax = plt.axes()
sn.heatmap(df_cm, annot=True,cmap="YlGnBu")

ax.set_title('Title')
plt.show()
```
